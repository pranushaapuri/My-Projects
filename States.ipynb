```python
import altair as alt
import pandas as pd
import numpy as np
import streamlit as st
from vega_datasets import data
```

```python
states = alt.topo_feature(data.us_10m.url, 'states')
loss_data = pd.read_csv ('losses2015_transformed.csv')
loss_data['AmountinMillions']=  (loss_data['Amount']/1000000)
```

```python

df =loss_data.groupby(['State_Abv','State_Code']).sum()
df= df.drop(['Damage_Code'],axis=1)
df.head(10)
```

```python
df = df.reset_index()
df.head(10)
```

```python

highlight = alt.selection_multi(fields=['State_Abv'])


map1=alt.Chart(states).mark_geoshape(stroke='white').encode(
    tooltip=['State_Abv:O','Amount:N'],
     opacity=alt.condition(highlight,alt.value(1),alt.value(0.2)),
     color='Amount:O'
#        color=alt.condition(highlight, alt.value('steelblue'), alt.value('gray')),
    
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(df, 'State_Code', ['State_Abv','Amount'])
).add_selection(highlight
).project(
    type='albersUsa'
).properties(
    width=400,
    height=300, title='Agricultural Losses by Cause in 2015'
)


bars=alt.Chart(loss_data).mark_bar().encode(
     y=alt.Y('Damage_Descp:N', axis=alt.Axis( title='Cause of Loss'),sort='-x'),
    x=alt.X('AmountinMillions:Q', axis=alt.Axis(title='Total Losses[$mil]')),
    order= alt.Order(
      # Sort the segments of the bars by this field
      'Amount',
      sort='descending'
    )
).transform_filter(highlight
    ).properties(
             width=400, height=300  
)



 

map1 | bars

```
